```python
transcript_url = "https://transcripts.fandom.com/wiki/"
show = "SpongeBob_SquarePants"

#urllib.request.urlretrieve(transcript_url+show, "transcript_list_url.txt")

outputFile = codecs.open("transcript_list_url.txt", "r", "utf8")
filestring = outputFile.read()

episode_list = re.findall('(?s)<h3><span class=\"mw-headline\" id=\"Season_1_.281999-2000.29\"> <u>Season 1 <small>(.*?)Movies', filestring, re.IGNORECASE)[0]
episode_list = re.findall('<a href="/wiki/(.*?)" title=".*?">(.*?)</a>', episode_list, re.IGNORECASE)
```

```python
#if(not(os.path.isdir(show+"_Episodes"))):
#    os.mkdir(show+"_Episodes")
#
#for episode in episode_list:
#    urllib.request.urlretrieve(transcript_url+episode[0], show+"_Episodes/"+episode[1]+".txt")
```

```python
transcripts = {}
categories = {}

for episode in episode_list:
    outputFile = codecs.open(show+"_Episodes/"+episode[1]+".txt", "r", "utf8")
    filestring = outputFile.read()
    
    #Get main transcript
    transcript = re.sub('(?s)<.*?>','',re.findall('(?s)<!-- End Google Tag Manager \(noscript\) -->(.*?)<!--', filestring, re.IGNORECASE)[0])
    try:
        transcripts[episode[1]] = max(*re.findall('(?s)^.*?Transcript(.*?)$|^.*?Plot(.*?)$|^.*?Script(.*?)$',transcript)[0])
    except:
        transcripts[episode[1]] = transcript
    
    #Get categories
    categories[episode[1]] = re.sub('"','',re.findall('(?s)wgCategories=\[(.*?)\]',filestring)[0]).split(',')
```

```python
completed_episodes = {name: transcripts[name] for name, category in categories.items() if 'Complete' in category}
```

```python
word_count = {episode: len(re.sub('\\n',' ',transcripts[episode]).split()) for episode in completed_episodes.keys()}
```

```python
#plt.hist(list(word_count.values()),bins='auto')
```

```python
sorted_episodes = sorted([[key,value] for key,value in word_count.items()],key=lambda x: x[1])
sorted_episodes
```

```python
filtered_episodes = {episode[0]: re.sub('\\\'','',re.sub('\\n',' ',re.sub('(?s)\\n[\w\s\.]*?:',' ',transcripts[episode[0]]))) for episode in sorted_episodes[:-3]}
```

```python
all_episodes = ' '.join(list(filtered_episodes.values()))
text_file = open('all_episodes.txt','w')
n = text_file.write(all_episodes)
text_file.close()
```
